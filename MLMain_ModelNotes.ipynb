# MODEL KURULUMU

```python
y_train = df_aggregate_month_main["TARGET_CAT"]
X_train = df_aggregate_month_main[["DOB_CAT","amt_pre_month1_cat","amt_pre_month2_cat","sex_F","sex_M"]]

y_test = df_aggregate_month_main_test["TARGET_CAT"]
X_test = df_aggregate_month_main_test[["DOB_CAT","amt_pre_month1_cat","amt_pre_month2_cat","sex_F","sex_M"]]

# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
knn = KNeighborsClassifier()
np.arange(1,50)
knn_params = {"n_neighbors": np.arange(1,50)}
knn_cv_model = GridSearchCV(knn, knn_params, cv = 10).fit(X_train, y_train)
knn_cv_model.best_score_
```

```python
knn_cv_model.best_params_
```

```python
knn_tuned = KNeighborsClassifier(n_neighbors = 41).fit(X_train, y_train)
```

```python
y_pred = knn_tuned.predict(X_test)
accuracy_score(y_test, y_pred)
```

# cart

```python
cart = DecisionTreeClassifier()
cart_params = {"max_depth": [1,3,5,8,10],
              "min_samples_split": [2,3,5,10,20,50]}
cart_cv_model = GridSearchCV(cart, cart_params, cv = 10, n_jobs = -1, verbose =2).fit(X_train, y_train)
cart_cv_model.best_params_
```

```python
cart_tuned = DecisionTreeClassifier(max_depth = 5, min_samples_split = 2).fit(X_train, y_train)
y_pred = cart_tuned.predict(X_test)
accuracy_score(y_test, y_pred)
```

# random forest

```python
rf = RandomForestClassifier()
rf_params = {"n_estimators": [100,200,500,1000],
            "max_features": [3,5,7,8],
            "min_samples_split": [2,5,10,20]}
rf_cv_model = GridSearchCV(rf, rf_params, cv = 10, n_jobs = -1, verbose = 2).fit(X_train, y_train)
rf_cv_model.best_params_
```

```python
rf_tuned = RandomForestClassifier(max_features = 5, 
                                  min_samples_split = 20, 
                                  n_estimators = 500).fit(X_train, y_train)
y_pred = rf_tuned.predict(X_test)
accuracy_score(y_test, y_pred)
```
