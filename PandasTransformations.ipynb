```python
# En son Pazartesi Gününü Tutacak bir kolon yarattım.weeknum dan gidilince bir weeknum iki aya birden girebiliyor. 
# Daha sonra ay hesanını da bu pazartesiye göre yaptım.
# Yani haftalık hesabı ilgili haftanın pazartesi gününe yazıyor oluyoruz.
# Aylık hesabı da pazartesi'lerin dahil olduğu aya göre yapıyoruz.

df_union_all["MondayDate"]=df_union_all["Dates_Date"] - pd.TimedeltaIndex(df_union_all["Dates_Date"].dt.dayofweek,unit='d') 
df_union_all['MondayDate_YYYY-MM']=df_union_all["MondayDate"].dt.strftime('%Y%m')
df_union_all
```

```python
# df_union_all.to_excel(
# r'timesheet-dshbrddata.xlsx',index=False,header=True
# )

df_union_all.to_excel(
r'df_union_all.xlsx',index=False,header=True
)
df_union_all.dtypes
```

# UTILAZATION HESAPLARI

```python
df_utl_and_hol_1=df_union_all.assign(
 uti_values=np.where(df_union_all['UTI_FLAG']=='UTI',df_union_all.Values,0),
 hol_values=np.where(df_union_all['UTI_FLAG']=='HOL',df_union_all.Values,0)
).groupby(['Resource','GRADE_CALC_CHAR','MondayDate_YYYY-MM','MondayDate']).agg({'uti_values':sum,'hol_values':sum,'Values':sum}).reset_index()

df_utl_and_hol=df_utl_and_hol_1[df_utl_and_hol_1['Values']!=0].copy()
```

```python
df_utl_and_hol["std_weekly_hours"]=40
df_utl_and_hol["net_hours"]=df_utl_and_hol["std_weekly_hours"]-df_utl_and_hol["hol_values"]
df_utl_and_hol["UTI_PERCENTAGE"]=(df_utl_and_hol["uti_values"]/df_utl_and_hol["net_hours"])*100
df_utl_and_hol["UTI_PERCENTAGE"] = df_utl_and_hol["UTI_PERCENTAGE"].fillna(0)
df_utl_and_hol
```

```python
df_utl_and_hol.to_excel(
r'df_utl_and_hol.xlsx',index=False,header=True
)
```

```python
df_utl_report=df_utl_and_hol.groupby(['GRADE_CALC_CHAR','MondayDate_YYYY-MM']).agg({'uti_values':sum,'net_hours':sum}).reset_index()

df_utl_report["UTI_PERCENTAGE"]=(df_utl_report["uti_values"]/df_utl_report["net_hours"])*100
df_utl_report["UTI_PERCENTAGE"] = df_utl_report["UTI_PERCENTAGE"].fillna(0)

df_utl_report
```

```python
df_utl_report.to_excel(
r'df_utl_report.xlsx',index=False,header=True
)
```

```python
df_utl_report_res=df_utl_and_hol.groupby(['Resource','MondayDate_YYYY-MM']).agg({'uti_values':sum,'net_hours':sum}).reset_index()

df_utl_report_res["UTI_PERCENTAGE"]=(df_utl_report_res["uti_values"]/df_utl_report_res["net_hours"])*100
df_utl_report_res["UTI_PERCENTAGE"] = df_utl_report_res["UTI_PERCENTAGE"].fillna(0)

df_utl_report_res


```

```python
df_utl_report_res.to_excel(
r'df_utl_report_res.xlsx',index=False,header=True
)
```
