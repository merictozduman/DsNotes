```python
df_top5=pd.DataFrame( df_forecasts2.sum().sort_values(ascending=[True]).iloc[:5].index,columns=["Models"])
df_top5["MODEL_ID"]=df_top5["Models"].str.replace('Forecast','')
df_top5["MODEL_ID"]=df_top5["MODEL_ID"].str.replace('x','')
df_top5["MODEL_ID"] = pd.to_numeric(df_top5["MODEL_ID"], errors='ignore')
df_top5
```

```python
import time
import numpy as np
from datetime import datetime, timedelta, date, time

TableStatsdict ={'Tarih':[],'Deger':[]}
max_date=dataset["Tarih"].max()+pd.DateOffset(1)


for i in range(12):
    TableStatsdict["Tarih"].append(max_date+pd.DateOffset(i))   
    TableStatsdict["Deger"].append(np.nan)                                   



Dataset_Fcast=pd.DataFrame.from_dict(TableStatsdict)
Dataset_Fcast
```

```python
dataset=dataset.append(Dataset_Fcast,ignore_index=True)
dataset.tail(25)
```

```python
ModelDict_prod ={'Model':[],'Fitted':[],'Model_Name':[],'Forecast':[]}

i=0

for row in df_top5.itertuples():  
    ModelDict_prod["Model"].append(HWES(seri.values,trend="add",seasonal_periods=row.MODEL_ID,seasonal="mul"))
    ModelDict_prod["Fitted"].append(ModelDict_prod["Model"][i].fit(optimized=True))
    ModelDict_prod["Model_Name"].append("Model"+str(i+1))
    ModelDict_prod["Forecast"].append(ModelDict_prod["Fitted"][i].forecast(steps=12))
    i=i+1

ModelDict_prod    

ModelPredictDict2 ={'Model':[],'Forecast':[],'Tarih':[]}        

# DataFrame'e atmak için ModelPredictDict'den forecast'leri sırayla çekiyorum
for i in range(5): 
    for j in range(0,12): 
      ModelPredictDict2["Model"].append(ModelDict_prod["Model_Name"][i])
      ModelPredictDict2["Forecast"].append(ModelDict_prod["Forecast"][i][j])
      ModelPredictDict2["Tarih"].append(Dataset_Fcast.loc[j,"Tarih"]) 



df_forecasts_prod=pd.DataFrame.from_dict(ModelPredictDict2 )
df_forecasts_prod
```

```python
df_forecasts_prod.pivot(index='Tarih',columns='Model', values='Forecast') 
```

```python
df_total=pd.merge(dataset, df_forecasts_prod.pivot(index='Tarih',columns='Model', values='Forecast') , on='Tarih', how='left')

df_total.iloc[-12:]
```
