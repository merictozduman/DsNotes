```python
import pandas as pd 

query = '''SELECT * FROM [AdventureWorksDW2012].[dbo].[FactInternetSales]''';
dataset = pd.read_sql(query, MsSqlCon)
dataset.head()
```

```python
dataset["OrderDateYear"]= pd.DatetimeIndex(dataset["OrderDate"]).year
dataset["OrderDateQuarter"]= pd.DatetimeIndex(dataset["OrderDate"]).quarter

dataset
```

```python
# Her yÄ±l iÃ§inde quarter karÅŸÄ±laÅŸtÄ±rmasÄ± yapmak iÃ§in, tÃ¼m quarterlarÄ± iÃ§eren yÄ±llarÄ± ayÄ±rÄ±yorum
dataset2011=dataset.loc[dataset['OrderDateYear']==2011]
# select distinct anlamÄ±nda:
# dataset2011.OrderDateYear.unique()

dataset2012=dataset.loc[dataset['OrderDateYear']==2012]
dataset2013=dataset.loc[dataset['OrderDateYear']==2013]

# TableStatsdict = {'Student Name': 'Berry', 'Roll No.': 12, 'Subject': 'English'}
```

```python
from scipy.stats import kruskal

# compare samples
stat, p = kruskal(dataset2011["SalesAmount"], dataset2012["SalesAmount"], dataset2013["SalesAmount"])
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05

if p > alpha:
 print('Same distributions (fail to reject H0)')
else:
 print('Different distributions (reject H0)')

```

```python
TableStatsdict = {'Test': ['Kruskal'], 'P_Value': [p], 'Stat_Value': [stat],'Is_Significiant':[p<alpha],'Object_Name':['KruskalByYear']}
print(TableStatsdict)
```

```python
from scipy.stats import kruskal

# compare samples
stat, p = kruskal(
    dataset2011["SalesAmount"].loc[dataset2011['OrderDateQuarter']==1], 
    dataset2012["SalesAmount"].loc[dataset2012['OrderDateQuarter']==1], 
    dataset2013["SalesAmount"].loc[dataset2013['OrderDateQuarter']==1]
     )
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
 print('Same distributions (fail to reject H0)')
else:
 print('Different distributions (reject H0)')
```

```python
# myd.add('apples',6)
TableStatsdict["Test"].append("Kruskal")
TableStatsdict["P_Value"].append(p)
TableStatsdict["Stat_Value"].append(stat)
TableStatsdict["Is_Significiant"].append((p < alpha))
TableStatsdict["Object_Name"].append("Q1Tests")
print(TableStatsdict)
```
